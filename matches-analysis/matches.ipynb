```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('matches.csv')
```

Q1. Extract the frequency of most man of the match award

```python
# Assuming the column name for "Man of the Match" is 'player_of_match'
# Get the frequency of each player
man_of_match_counts = df['player_of_match'].value_counts()

# Get the player with the most "Man of the Match" awards
most_man_of_match_player = man_of_match_counts.idxmax()
most_man_of_match_count = man_of_match_counts.max()

print(f"The player with the most 'Man of the Match' awards is '{most_man_of_match_player}' with '{most_man_of_match_count}' awards.")
```

Q2. Extract the top 10 players with most man of the match award 

```python
top_10_man_of_match_players = man_of_match_counts.head(10)

print("Top 10 players with the most 'Man of the Match' awards:")
print(top_10_man_of_match_players)
```

Q3. Extract top 5 players with most man of the match award

```python
top_5_man_of_match_players = man_of_match_counts.head(5)

print("Top 10 players with the most 'Man of the Match' awards:")
print(top_5_man_of_match_players)
```

Q4. Draw bar plot of top 5 most man of the match award winning player

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
top_5_man_of_match_players.plot(kind='bar', color='skyblue')
plt.title('Top 5 Players with Most Man of the Match Awards')
plt.xlabel('Players')
plt.ylabel('Number of Awards')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Q5. Draw violin plot of top 5 most man of the match award winning player

```python
plt.violinplot(top_5_man_of_match_players)
plt.title('Top 5 Players with Most Man of the Match Awards')
plt.xlabel('Players')
plt.ylabel('Number of Awards')
plt.grid(True)
plt.show()
```

Q6. Evaluate the frequency of result column

```python
result_counts = df['result'].value_counts()

print("Frequency of each result:")
print(result_counts)
```

Q7. Find out the number of toss win with respect to each team

```python
toss_win_counts = df['toss_winner'].value_counts()

print("Number of toss wins for each team:")
print(toss_win_counts)
```

Q8. Extract the record player a team win batting first

```python
bat_decision_matches = df[df['toss_decision'] == 'bat']

bat_and_win_matches = bat_decision_matches[bat_decision_matches['toss_winner'] == bat_decision_matches['winner']]

print("Matches where the toss decision was 'bat' and the team that decided to bat won the match:")
print(bat_and_win_matches)
```

Q9. Make the histogram of win by run column

```python
plt.hist(df['result_margin'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Win by Run')
plt.xlabel('Win by Run')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Q10. Find out the number of wins with respect to each team after batting first

```python
batting_first_wins = df[(df['toss_decision'] == 'bat') & (df['toss_winner'] == df['winner'])]
wins_by_team = batting_first_wins['winner'].value_counts()
print(wins_by_team)
```

Q11. Make the bar plot for top 3 teams with most wins after batting first

```python
top_3_teams = wins_by_team.head(3)

# Plot the bar chart
plt.figure(figsize=(8, 6))
top_3_teams.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 3 Teams with Most Wins after Batting First')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Q16. Calculate the average winning margin (in runs) for all teams when they bat first.

```python
batting_first_wins = df[df['result'] == 'runs']

average_winning_margin = batting_first_wins['result_margin'].mean()

print("Average winning margin (in runs) for all teams when they bat first:")
print(average_winning_margin)
```

Q17. Determine which team has lost the most matches when batting first.

```python
batting_first_losses = df[(df['toss_decision'] == 'bat') & (df['result'] == 'wickets')]

losses_by_team_batting_first = batting_first_losses['team1'].value_counts() + batting_first_losses['team2'].value_counts()

team_with_most_losses = losses_by_team_batting_first.idxmax()
most_losses_count = losses_by_team_batting_first.max()

print(f"The team that has lost the most matches when batting first is {team_with_most_losses} with {most_losses_count} losses.")
```
